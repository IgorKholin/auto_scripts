```python
import pandas as pd
```

```python
# прописываем путь до анкеты
path_to_file = (r'C:\Users\Home\Documents\anket.xlsx')
path_to_second_file = (r'C:\Users\Home\Documents\anket2.xlsx') 

# функция по обработке файлика
def file_handler(path_to_file) -> pd.DataFrame:
    first_file = pd.read_excel(path_to_file, header=None)
    # выделям необходимые для нас поля
    code_of_tk = first_file.iloc[1, 1]
    cluster = first_file.iloc[1, 3]
    competitor = first_file.iloc[1, 13]
    # удаляем ненужные строки(первые две)
    df = first_file.iloc[2:].reset_index(drop=True)
    df.columns = df.iloc[0]
    df = df.iloc[1:]
    # считаем кол-во строк, чтобы корректно сопоставить новые столбцы с нашей табличкой
    length_data = len(df)
    
    # формируем списки из необходимых нам данных(код ТК, кластеры и т.д.)
    def forming_lists(measure, length_data) -> list:
        lst = []
        for _ in range(length_data):
            lst.append(measure)
        return lst

    # создаём непосредственно сами столбцы и добавляем их нашу таблицу
    list_clusters = forming_lists(code_of_tk, length_data)
    list_tk_code = forming_lists(cluster, length_data)
    list_competitor = forming_lists(competitor, length_data)
    df['Кластер'] = list_clusters
    df['Код_ТК'] = list_tk_code
    df['Конкурент'] = list_competitor
    
    return df

# теперь просто объединяем две наши таблицы и скачиваем их в необходимую папку
def concate_and_save(path_to_file, path_to_second_file) -> pd.DataFrame:
    anket = file_handler(path_to_file) # обработка анкеты
    anket2 = file_handler(path_to_second_file)
    result = pd.concat([anket, anket2], axis=0, ignore_index=True) # объединение
    result.to_excel(r"C:\Users\Home\Documents\file.xlsx", index=False)
    return result

concate_and_save(path_to_file, path_to_second_file)
```
